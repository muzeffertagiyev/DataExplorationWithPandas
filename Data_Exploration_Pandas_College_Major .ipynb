```python
# with the help of that function we will recreate our data without NANs
clean_df = df.dropna()
```

```python
# it gets maximum value in Starting Median Salary column
clean_df['Starting Median Salary'].max()
```

```python
# it gets index of maximum value in Starting Median Salary column
clean_df['Starting Median Salary'].idxmax()
```

```python
# they both get major value from  Undraduate Major which locates in index 43
clean_df['Undergraduate Major'].loc[43]
clean_df['Undergraduate Major'][43]
```

```python
# it gets all values in index 43
clean_df.loc[43]
```

```python
index_highest_salary_in_mid = clean_df['Mid-Career Median Salary'].idxmax()
highest_salary_in_mid = clean_df['Mid-Career Median Salary'].max()
major_in_highest_mid = clean_df['Undergraduate Major'].loc[index_highest_salary_in_mid]

print('highest salary-',highest_salary_in_mid)
print('major-',major_in_highest_mid)
```

```python
index_lowest_salary_in_start = clean_df['Starting Median Salary'].idxmin()
lowest_salary_in_start = clean_df['Starting Median Salary'].min()
major_in_lowest_in_start = clean_df['Undergraduate Major'].loc[index_lowest_salary_in_start]

print('Lowest salary - ',lowest_salary_in_start)
print('Major - ', major_in_lowest_in_start)
```

```python
index_lowest_salary_mid = clean_df['Mid-Career Median Salary'].idxmin()
lowest_salary_mid = clean_df['Mid-Career Median Salary'].min()
major_in_lowest_mid = clean_df['Undergraduate Major'].loc[index_lowest_salary_mid]

print('Lowest salary In Mid - ', lowest_salary_mid)
print('Major - ', major_in_lowest_mid)
```

```python
# it substracts two values(we can also use normal subtract) and then insert values after 
# subtraction into clean_df using insert function(1- is index where we want to inset ,'Spread' is column name,
# spread_col is values)
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
clean_df.insert(1,'Spread', spread_col)
clean_df.head()
```

```python
# this is sorting values as ascending by 'Spread' column and after that as you see below we can call 
# just two columns by using this method
low_risk = clean_df.sort_values('Spread', ascending=True)
low_risk[['Undergraduate Major','Spread']].head()
```

```python
# Task 1:Using the .sort_values() method, can you find the degrees with the highest potential? 
# Find the top 5 degrees with the highest values in the 90th percentile. 

print("Top 5 highest Mid-Career 90th Percentile salary")
degrees_with_highest_pote = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending = False)
degrees_with_highest_pote[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

```python
# Task 2: Also, find the degrees with the greatest spread in salaries. Which majors have the largest 
# difference between high and low earners after graduation.
print("Top 5 majors with greatest spread in salaries")
degrees_with_greatest_spread = clean_df.sort_values('Spread', ascending=False)
degrees_with_greatest_spread[['Undergraduate Major', 'Spread']].head()
```
